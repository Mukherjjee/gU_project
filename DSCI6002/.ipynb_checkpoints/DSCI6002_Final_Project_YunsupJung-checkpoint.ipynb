Q1. Which stock seems the best dividend return for next dividnd based on current price (largest percent_return_next_dividend value)?
>T is the higest dividied return company for next dividend.

```python
%%sql

SELECT stock, percent_return_next_dividend FROM dow_jones_index
GROUP bY stock
ORDER by percent_return_next_dividend DESC
LIMIT 3
```

```python
import numpy as np

percent_your_pick_go_up_next_week = 1
for i in range(5):
    percent_your_pick_go_up_next_week *= (go_up_stock-i)/(total_weeks-i)
print('The {0:.3}% chance to your all pick(all 5 stocks) go up in next week.'.format(percent_your_pick_go_up_next_week*100))
```

### Linear Regression Model

```python
import statsmodels as sm
import statsmodels.api as sms
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
from statsmodels.tsa.arima_model import ARMA

from statsmodels.stats.diagnostic import het_breushpagan
from statsmodels.graphics.correlation import plot_corr
from statsmodels.graphics.gofplots import qqplot
from statsmodels.graphics.regressionplots import plot_leverage_resid2

dow = eda_price.copy()

#make sure upper letter in columns, then replace upper letter to lower letter
dow.columns = dow.columns.map(lambda x: x.lower())
```

```python
#organize data set in order to apply multiple regression model
#add volume, price_up and dividend to columns to see whether x = mutiple variables effect y = stock price 
dow['volume'] = df.groupby(['stock']).volume.mean().values
dow['price_up'] = df.groupby(['stock']).up.sum().values
div_per = df.groupby(['stock']).percent_return_next_dividend.mean().values
dow['dividend'] = div_per * eda_price['Mean'].values
dow = dow.rename(columns={'mean':'price'})
dow = dow.set_index('stock')
dow.head(3)
```

```python
dow.corr()
```

```python
#make correlated map to see correlation between variables
sm.graphics.correlation.plot_corr(dow.corr(), xnames = dow.columns.tolist());
```

>The correlated matrix graph explains that std and dividend are highly correlated with price, and volume in least correlated with price. Also, I can see std and price_up are about 60% correlated with dividend, and price_up is about 40% correlated with std.

```python
#check the model fit in order to know which variables effect to price
X1 = dow[['std','volume','price_up','dividend']]
X1 = sms.add_constant(X1)
y = dow[['price']]
model1 = sms.OLS(y,X1)
results1 = model1.fit()
results1.summary()
```

>Although Adj. R-squared and Prob (F-statistic), p-value for whole model, tell a good model fit to predict stock price, the p-values of some variables such as price_up and volums are much greater than 0.05. 
<br>
Hence, I will take out these variable by decrease order; volume and price_up in order to get a good model 
which is fit to predict stock price.

```python
#remove variable volume which is not really effect to price based on p-value
X2 = X1.drop('volume', axis = 1)
model2 = sms.OLS(y,X2)
results2 = model2.fit()
results2.summary()
```

>In same reason above, I will remove variable price_up, then make new model until I got a good model fit.

```python
#remove variable price_up which is not really effect to price based on p-value
X3 = X2.drop('price_up', axis = 1)
model3 = sms.OLS(y,X3)
results3 = model3.fit()
results3.summary()
```

>Now, I got a great fit model for price by p-values of each variables (less than 0.05).
<br>
In addition, Adj. R-squared and Prob (F-statistic), p-value for whole model, tell a good model fit to predict stock price (values are improved as compare with above models).
The model is 88.5% of the variation in price.  
Although Adj. R-squared is almost same as previous models, F-statistic is much greater than original model, and p-value is smaller. 
<br>
Hence, it is better model for price.

* Interpret the coefficients of model.

>For a change of one unit in x=std, the value of y=price will increase by 1.847.
<br>
For a change of one unit in x=dividend, the value of y=price will increase by 0.134.

* Box-Jenkins Method
>General Steps:
<br>
>* Identify the steps needed to achieve stationarity (detrending, differencing, and/or transformations)
>* Utilize ACF to see if stationarity is achieved. Lags should die down quickly.
>* Use ACF/PACF to identify potential models by where the lags cut off.
>>* PACF identifies AR(p)
>>* ACF identifies MA(q)
>* Fit model and examine residuals
>>* If white noise, you're done
>>* Else, identify additional components and update model

```python
#make function to test acf and pacf
def acf_pacf(ts, lags):
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sms.graphics.tsa.plot_acf(ts, lags=lags, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sms.graphics.tsa.plot_pacf(ts, lags=lags, ax=ax2)
```

```python
#plot difference
ts_mean.diff(1).plot(figsize=(12,8));
```

```python
#ACF identifies MA(q)
#PACF identifies AR(p)
acf_pacf(ts_mean,24)
```

```python
#apply arma model; ACF identifies MA(1), PACF identifies AR(1)
arma_mod11 = ARMA(ts_mean, (1,1)).fit()
print(arma_mod11.params)
```

```python
arma_mod11.summary2()
```

>I got negative log likelihood -22.219, and the values of BIC(57.3135) and AIC(52.4380).
<br>
The p-value of MA(1) is litte high than 0.05, so it is not significant(dont keep it)
, and the p-value of  AR(1) is significant. 

```python
#try ARMA(2,2)
arma_mod22 = ARMA(ts_mean, (2,2)).fit()
arma_mod22.summary2()
```

>I got negative log likelihood -21.239, and the values of BIC(61.7922) and AIC(54.4789).

<br>
Based on values, this model is better than previous model.

### GLM(Generalized Linear Model)

```python
#Dow Jones index move in range +/-1.8 * standard deviation from Ha, so I make new column 'Risk Var'.
glm_dow = dow.copy()
glm_dow['risk_var'] = 1.8*glm_dow['std']*100/glm_dow['price']
glm_dow['risk_var'].describe()
```

```python
#petition group into risk or safe based on mean of risk_var; 0 is stable, 1 is aggressive
glm_dow['aggressive'] = glm_dow['risk_var'].apply(lambda x: 1 if x>= glm_dow['risk_var'].mean() else 0)
```

```python
#compare week1 and week25, then add column for +/- return at the eand of half year in 2011
ret = []
for i in com_list:
    name = df[df['stock']==i]
    week1 = name[name['week']==1].weekly_mean_price.values
    week25 = name[name['week']==25].weekly_mean_price.values
    ret.append(week25 - week1)
glm_dow['outcome'] = ret
```

```python
#add new column number of times price up 
glm_dow['up'] = df.groupby(['stock']).up.sum()
```

```python
#modify outcome column to glm; 0 is stock price went down, 1 is stock price went up at the end of first half year 2011.
glm_dow['outcome'] = glm_dow['outcome'].map(lambda x: 1 if x>0 else 0)
```

```python
#clean data set
glm_dow = glm_dow.drop(['price', 'std', 'dividend', 'risk_var'], axis=1)
glm_dow
```

```python
#plotting
glm_dow.hist(figsize=(6,6));
```

```python
#check the intercept
lm = ols("outcome ~ 1", data=glm_dow).fit()
lm.params
```

>Modeling the intercept tells that 60% of stock were still price moved up.

```python
#modeling aggressive group
lm1 = ols("outcome ~ aggressive", data=glm_dow).fit()
lm1.params
```

```python
#modeling aggressive group + number of weeks price moved up
#Linear Regression using OLS in statsmodels
lm2 = ols("outcome ~ aggressive + up", data=glm_dow).fit()
lm2.params
```

```python
#Linear Regression using GLM in statsmodels
glm_norm = smf.glm("outcome ~ aggressive + up", data=glm_dow).fit()
glm_norm.params
```

```python
#Linear Regression using GLM (with family specified)
glm_norm1 = smf.glm("outcome ~ aggressive + up", data=glm_dow, family=sms.families.Gaussian()).fit()
glm_norm1.params
```

```python
#plot linear regression
sns.lmplot("up", "outcome", glm_dow, hue="aggressive", size=8, aspect=1.5, y_jitter=.1,);
```

Since linear regression is not the appropriate model to use if the $Y$ variable is binary (0/1).
Therefore, I need The Logistic Regression Model.

* The Logistic Regression Model

```python
#make logistic regression model
logit_model1 = smf.logit("outcome ~ aggressive + up", data=glm_dow).fit()
print(logit_model1.params)
logit_model1.summary()
```

>The aggressive stock which price moved up in 14 weeks:
<br>
$-7.6271 + 0.1517 \times 1 + 0.6530 \times 14 = 1.667$

<br>
>It has a $166.7\%$ chance of being profit at the end of half year.
>This results are more make sense, so I choose the Logistic Regression Model.
